### 2.1 What proportion of the gunny bags have a breaking strength of less than 3.17 kg per sq cm?

Lets import necessary libararies to calclate the area under the curve


```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import scipy.stats as stats
from scipy.stats import norm
from scipy import stats
```

```python
# Given parameters
mean = 5.0  # Mean breaking strength (in kg per sq. cm)
std_dev = 1.5  # Standard deviation (in kg per sq. cm)
x = 3.17  # Breaking strength threshold (in kg per sq. cm)
```

```python
# To calclate the area towards left of Z score
# find the cumulative probability
# norm.cdf() calculates the cumulative probability
prob_less = norm.cdf(3.17, 5, 1.5)
print(
    "The proportion of the gunny bags have a breaking strength of less than 3.17 kg per sq cm is",
    round(prob_less, 4),
)
```

```python
# Visualize the normal distribution and shaded area
x_values = np.linspace(mean - 4 * std_dev, mean + 4 * std_dev, 1000)
y_values = norm.pdf(x_values, mean, std_dev)

plt.plot(x_values, y_values, 'b-', linewidth=2)
plt.fill_between(x_values, y_values, where=(x_values <= x), color='red', alpha=0.5)
plt.title('Breaking Strength of less than 3.17 kg per sq cm')
plt.xlabel('Breaking Strength (kg per sq. cm)')
plt.ylabel('Probability Density')
plt.axvline(x=x, color='black', linestyle='--')
plt.annotate(f'Proportion = {prob_less:.2f}', (x + 0.2, 0.02), fontsize=12, color='black')
plt.grid(True)
plt.show()

print(f"The proportion of gunny bags with a breaking strength less than {x} kg per sq. cm is {prob_less:.4f}")
```

### 2.3 What proportion of the gunny bags have a breaking strength between 5 and 5.5 kg per sq cm.?

```python
# Given parameters
mean = 5.0  # Mean breaking strength (in kg per sq. cm)
std_dev = 1.5  # Standard deviation (in kg per sq. cm)
lower_limit = 5.0  # Lower breaking strength threshold (in kg per sq. cm)
upper_limit = 5.5  # Upper breaking strength threshold (in kg per sq. cm)
```

```python
# Calculate the z-scores for lower and upper limits
z_lower = (lower_limit - mean) / std_dev
z_upper = (upper_limit - mean) / std_dev
```

```python
# Calculate the proportion between the limits using the CDF
proportion_between_limits = norm.cdf(z_upper) - norm.cdf(z_lower)
proportion_between_limits
```

```python
# Visualize the normal distribution and shaded area
x_values = np.linspace(mean - 4 * std_dev, mean + 4 * std_dev, 1000)
y_values = norm.pdf(x_values, mean, std_dev)

plt.plot(x_values, y_values, 'b-', linewidth=2)
plt.fill_between(x_values, y_values, where=(x_values >= lower_limit) & (x_values <= upper_limit), color='purple', alpha=0.5)
plt.title('Breaking Strength between 5 and 5.5 kg per sq cm')
plt.xlabel('Breaking Strength (kg per sq. cm)')
plt.ylabel('Probability Density')
plt.axvline(x=lower_limit, color='black', linestyle='--')
plt.axvline(x=upper_limit, color='black', linestyle='--')
plt.annotate(f'Proportion = {proportion_between_limits:.2f}', (lower_limit + 0.2, 0.02), fontsize=12, color='black')
plt.grid(True)
plt.show()

print(f"The proportion of gunny bags with a breaking strength between {lower_limit} and {upper_limit} kg per sq. cm is {proportion_between_limits:.4f}")
```

### 3.1 Zingaro has reason to believe that the unpolished stones may not be suitable for printing. Do you think Zingaro is justified in thinking so?

To determine if Zingaro has reason to believe that the unpolished stones may not be suitable for printing, we can perform a hypothesis test using the given data. In this case, we'll use a hypothesis test for the population mean with the following hypotheses:

Null Hypothesis (H0): The mean Brinell's hardness index of unpolished stones () is greater than or equal to 150 (  150).

Alternative Hypothesis (H1): The mean Brinell's hardness index of unpolished stones () is less than 150 ( < 150).

We will perform a one-sample t-test to determine if there is enough evidence to reject the null hypothesis in favor of the alternative hypothesis.

```python
# Given data (Brinell's hardness index of unpolished stones)
hardness_values = zing['Unpolished']
```

```python
# Set the significance level (alpha)
alpha = 0.05
```

```python
# Define the null and alternative hypotheses
null_hypothesis = "The mean Brinell's hardness index of unpolished stones is greater than or equal to 150."
alternative_hypothesis = "The mean Brinell's hardness index of unpolished stones is less than 150."
```

```python
# Calculate the sample mean and standard error of the mean (SEM)
sample_mean = np.mean(hardness_values)
sample_std = np.std(hardness_values, ddof=1)  # ddof=1 for sample standard deviation
sample_size = len(hardness_values)
sem = sample_std / np.sqrt(sample_size)
```

```python
zing['Unpolished']
```

```python
zing['Treated and Polished']
```

```python
# Calculate the t-statistic
t_statistic = (sample_mean - 150) / sem
t_statistic
```

```python
# Calculate the degrees of freedom
degrees_of_freedom = sample_size - 1
degrees_of_freedom
```

```python
# Calculate the p-value
p_value = stats.t.cdf(t_statistic, df=degrees_of_freedom)
p_value
```

```python
# Determine whether to reject the null hypothesis
reject_null = p_value < alpha
reject_null
```

```python
# Print the results
print("Null Hypothesis:", null_hypothesis)
print("Alternative Hypothesis:", alternative_hypothesis)
print(f"Sample Mean: {sample_mean:.2f}")
print(f"t-statistic: {t_statistic:.2f}")
print(f"P-value: {p_value:.4f}")
print("Significance Level (alpha):", alpha)

if reject_null:
    print("Conclusion: Reject the null hypothesis.")
    print("Zingaro has reason to believe that the unpolished stones may not be suitable for printing.")
else:
    print("Conclusion: Fail to reject the null hypothesis.")
    print("There is not enough evidence to suggest that the unpolished stones are unsuitable for printing.")
```

### 3.2 Is the mean hardness of the polished and unpolished stones the same?

To determine if the mean hardness of the polished and unpolished stones is the same, I have performed a two-sample t-test. I
will compare the means of the two samples (polished and unpolished stones) and test whether they are significantly different from each other.

Null Hypothesis (H0): The mean hardness of polished stones is the same as the mean hardness of unpolished stones.

Alternative Hypothesis (H1): The mean hardness of polished stones is different from the mean hardness of unpolished stones.

```python
# Given data (Brinell's hardness index of polished and unpolished stones)
polished_hardness = np.array(zing['Treated and Polished'])
unpolished_hardness = np.array(zing['Unpolished'])
```

```python
# Set the significance level (alpha)
alpha = 0.05
```

```python
# Define the null and alternative hypotheses
null_hypothesis = "The mean hardness of polished stones is the same as the mean hardness of unpolished stones."
alternative_hypothesis = "The mean hardness of polished stones is different from the mean hardness of unpolished stones."
```

```python
# Perform a two-sample t-test
t_statistic, p_value = stats.ttest_ind(polished_hardness, unpolished_hardness, equal_var=False)
p_value
```

```python
# Determine whether to reject the null hypothesis
reject_null = p_value < alpha
reject_null
```

```python
# Print the results
print("Null Hypothesis:", null_hypothesis)
print("Alternative Hypothesis:", alternative_hypothesis)
print(f"t-statistic: {t_statistic:.2f}")
print(f"P-value: {p_value:.4f}")
print("Significance Level (alpha):", alpha)
if reject_null:
    print("Conclusion: Reject the null hypothesis.")
    print("The mean hardness of polished and unpolished stones is different.")
else:
    print("Conclusion: Fail to reject the null hypothesis.")
    print("There is not enough evidence to suggest that the mean hardness of polished and unpolished stones is different.")
```

Now, the normality and equality of variance assumptions need to be checked.

For testing of normality, Shapiro-Wilks test is applied to the response variable (Response)

For equality of variance, Levene test is applied to the response variable (Response)

### Shapiro-Wilks test for Dentist
We will test the null hypothesis

H0:
 Dentist follows a normal distribution against the alternative hypothesis
 
Ha:
  Dentist does not follow a normal distribution.

```python
# Assumption 1: Normality
# import the required function
from scipy import stats

# find the p-value
w, p_value = stats.shapiro(dental1['Dentist']) 
print('The p-value is', p_value)
```

Since p-value of the test is very small, we fail to reject the null hypothesis that the response follows the normal distribution. In other words it does not follow a normal distribution.

```python
# Levene's test for homogeneity of variance
levene_stat, levene_p_value = levene(
    dental1['Dentist'][dental1['Dentist'] == 1],
    dental1['Dentist'][dental1['Dentist'] == 2],
    dental1['Dentist'][dental1['Dentist'] == 3],
    dental1['Dentist'][dental1['Dentist'] == 4],
    dental1['Dentist'][dental1['Dentist'] == 5]
)

# Set the significance level (alpha)
alpha = 0.05
```

```python
# Print the results and conclusions
print("Levene's Test for Homogeneity of Variance for Dentist Variable:")
print("Test Statistic:", levene_stat)
print("p-value:", levene_p_value)

if levene_p_value > alpha:
    print("Conclusion: The variances of the Dentist variable in different groups are homogeneous.")
else:
    print("Conclusion: The variances of the Dentist variable in different groups are not homogeneous.")
```

### Let's test whether the assumptions are satisfied or not

The populations are normally distributed - No, the normality assumption is verified using the Shapiro-Wilks test.

Population variances are equal - No, the homogeneity of variance assumption is verified using the Levene's test.

### 4.4 How does the hardness of implants vary depending on dentists and methods together?

We will first state the null and alternate hypotheses, check the assumptions, conduct the test, and then write conclusions:

### Step 1: State the null and alternate hypotheses:

For the two-way ANOVA, 

The null hypothesis (H0) is that there is no significant difference in the means between groups based on Dentist and Method, and 

The alternate hypothesis (Ha) is that there is a significant difference.

### Step 2: Check the assumptions:

Before conducting the two-way ANOVA, you should check the following assumptions:

Normality: Check the normality of the residuals for each combination of Dentist and Method. You can use the Shapiro-Wilk test for this.

Homogeneity of Variance: Check if the variances are homogeneous for each combination of Dentist and Method. You can use Levene's test for this.

To check the assumptions of a two-way ANOVA, we performed normality tests and homogeneity of variance tests for each combination of Dentist and Method. Below is  the code using the Shapiro-Wilk test for normality and Levene's test for homogeneity of variance:

```python
# Check normality of residuals for each combination
for dent in dental1['Dentist'].unique():
    for method in dental1['Method'].unique():
        subset = dental1[(dental1['Dentist'] == dent) & (dental1['Method'] == method)]['Response']
        shapiro_test = stats.shapiro(subset)
        print(f"Dentist {dent}, Method {method} Shapiro-Wilk p-value: {shapiro_test[1]}")
```

### Insights

From the above Shapiro's Test of Normality. We can see that only Dentist1 and Method1 2.4% is having p-value less than the level of significant 5% which means that only this combination is not normal distribution. The rest of the combinations of Dentists & Methods are normally distributed.




```python
# Check homogeneity of variances for each combination
for dent in dental1['Dentist'].unique():
    for method in dental1['Method'].unique():
        subset = dental1[(dental1['Dentist'] == dent) & (dental1['Method'] == method)]['Response']
        if len(subset) >= 2:
            levene_test = stats.levene(*[subset for _ in range(len(subset))])
            print(f"Dentist {dent}, Method {method} Levene's p-value: {levene_test[1]}")
```

### Insights

Levene's test is used to check whether the variances of different groups are approximately equal.

The p-value is less than the significance level (typically 0.05), we reject the null hypothesis (H0), suggesting that the variances are not equal (heteroscedasticity).

### Step 3: Conduct the two-way ANOVA and compute the p-value:

Null Hypothesis (H0): There are no significant differences in the means of the response variable among different Dentists, Methods, or their interactions.

Alternative Hypothesis (H1): At least one of the factors (Dentist, Method) or their interaction has a significant effect on the means of the response variable.

```python
# Fit the two-way ANOVA model
model = ols('Response ~ C(Dentist) * C(Method)', data=dental1).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
print(anova_table)
```

### Insights

Analyzed the effects of two categorical factors (Dentist and Method) on a continuous response variable (Hardness of dental implants). 

p-value is less than your chosen significance level (e.g., 0.05), we reject the null hypothesis (H0). This indicates that there are statistically significant differences in the means of the response variable among at least one combination of Dentist, Method, or their interaction.

We have evidence to suggest that at least one of these factors has a significant effect on the response variable.

### Step 5: Identify which Dentists and Methods combinations are different, and which interaction levels are different:

If the two-way ANOVA indicates significant differences, you can perform post-hoc tests to identify which specific combinations of Dentists and Methods are different from each other.

In Step 5, you can perform post-hoc tests to identify which Dentists and Methods combinations are different from each other, and which interaction levels are different. One common post-hoc test for this purpose is the Tukey HSD (Honestly Significant Difference) test. 

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd
from statsmodels.stats.multicomp import MultiComparison
```
