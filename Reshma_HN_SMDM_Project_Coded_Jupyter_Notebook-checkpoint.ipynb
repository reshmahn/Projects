```python
## We can treat the outliers using percentile  or with Inter quartile Range
## Here, we are using percentile to replace outliers

min_Total_salary, max_total_salary = austo['Total_salary'].quantile([0.05,0.95])
print(min_Total_salary)
print(round(max_total_salary,2))
```

```python
austo["Total_salary"] = np.where(austo["Total_salary"] <  min_Total_salary, min_Total_salary, austo["Total_salary"])
austo["Total_salary"] = np.where(austo["Total_salary"] >  max_total_salary, max_total_salary, austo["Total_salary"])
```

```python
sns.boxplot(x="Total_salary", data=austo)
plt.show()
```
